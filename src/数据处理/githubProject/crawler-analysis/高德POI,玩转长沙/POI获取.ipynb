### 简单调用

```python
import requests
import json
import pprint
```

```python
url = 'https://restapi.amap.com/v3/place/text?types=050000&city=长沙&offset=20&page=1&key=这是你的key'

res = requests.get(url)
print(res.text)
```

### 获取数据

```python
result = json.loads(res.text)
pprint.pprint(result)
```

```python
count = result['count']
count
```

```python
pois = result['pois']
for poi in pois:
    name = poi['name']
    types = poi['type']
    type1 = types.split(';')[0]
    type2 = types.split(';')[1]
    type3 = types.split(';')[2]
    location = poi['location']
    lon = location.split(',')[0]
    lat = location.split(',')[1]
    print(name,type1,type2,type3,lon,lat)
```

### 餐饮数据爬取

```python
import requests
import json
import csv
```

```python
fp = open('poi.csv','a+',newline='',encoding='utf-8')
writer = csv.writer(fp)
writer.writerow(['name','typ1','typ2','typ3','lon','lat'])

urls = ['https://restapi.amap.com/v3/place/text?types=050000&city=长沙&offset=20&page={}&key=这是你的key'.format(str(i)) for i in range(1,100)]

for url in urls:
    res = requests.get(url)

    result = json.loads(res.text)
    count = result['count']
    if count == '0':
        break
    pois = result['pois']
    for poi in pois:
        name = poi['name']
        types = poi['type']
        type1 = types.split(';')[0]
        type2 = types.split(';')[1]
        type3 = types.split(';')[2]
        location = poi['location']
        lon = location.split(',')[0]
        lat = location.split(',')[1]
        print(name,type1,type2,type3,lon,lat)
        writer.writerow([name,type1,type2,type3,lon,lat])
```
