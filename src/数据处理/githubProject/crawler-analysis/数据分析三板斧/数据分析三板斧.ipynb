### 处理数据

```python
data['Brand'].unique()
```

```python
data = data[data['Brand'] == 'Starbucks']
```

```python
data['Brand'].unique()
```

```python
data.isnull().sum()
```

```python
data[data['City'].isnull()]
```

```python
def fill_na(x):
    return x

data['City'] = data['City'].fillna(fill_na(data['State/Province']))
data[data['Country']=='EG']
```

```python
data['Country'][data['Country'] == 'TW'] = 'CN'
```

```python
data[data['Country'] == 'TW']
```

### 数据分析+可视化

```python
country_count = data['Country'].value_counts()[0:10]
country_count
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['simhei'] #指定默认字体
plt.rcParams['axes.unicode_minus'] = False   #解决保存图像是负号'-'显示为方块的问题
%matplotlib inline

country_count.plot(kind='bar')
```
