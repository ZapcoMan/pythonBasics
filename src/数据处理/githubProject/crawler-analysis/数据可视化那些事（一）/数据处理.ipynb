```python
data = data[data['Brand'] == 'Starbucks']
```

```python
data.isnull().sum()
```

```python
data[data['City'].isnull()]
```

```python
def fill_na(x):
    return x

data['City'] = data['City'].fillna(fill_na(data['State/Province']))
data[data['Country']=='EG']
```

```python
data['Country'][data['Country'] == 'TW'] = 'CN'
```

```python
data[data['Country'] == 'TW']
```

### 数据分析+可视化

```python
country_count = data['Country'].value_counts()[0:10]
country_count
```

```python
writer = pd.ExcelWriter('可视化表格.xlsx')
country_count.to_excel(writer,sheet_name='国家分布前十')
```

```python
city_count = data['City'].value_counts()[0:10]
city_count
```

```python
city_count.to_excel(writer,sheet_name='城市分布前十')
```

```python
china_data = data[data['Country'] == 'CN']
china_data.head()
```

```python
city_count = china_data['City'].value_counts()[0:10]
city_count
```

```python
city_count.to_excel(writer,sheet_name='中国城市分布前十')
```
