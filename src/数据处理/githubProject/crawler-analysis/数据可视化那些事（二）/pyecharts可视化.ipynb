```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('friend.csv',encoding='utf-8')
data.head()
```

### 基本使用

```python
sex = data.groupby('Sex')['Sex'].count()
sex
```

```python
from pyecharts import Pie

pie = Pie("男女分布", title_pos='center')
pie.add(
    "",
    ['外星人','男性','女性'],
    list(sex),
    radius=[40, 75],
    label_text_color=None,
    is_label_show=True,
    legend_orient="vertical",
    legend_pos="left",
)
pie
```

```python
new_data = data[data['Signature'].notnull()]
new_data.head()
```

```python
str_data = ''
for i in range(new_data.shape[0]):
    str_data = str_data + new_data.iloc[i,4]
str_data
```

```python
import re
str_data = re.sub('span', '',str_data,re.S)
str_data = re.sub('class', '',str_data,re.S)
str_data = re.sub('emoji', '',str_data,re.S)
str_data
```

```python
import jieba.analyse
tags = jieba.analyse.extract_tags(str_data, topK=50, withWeight=True)
```

```python
label = []
attr = []
for item in tags:
    print(item[0]+'\t'+str(int(item[1]*1000)))
    label.append(item[0])
    attr.append(int(item[1]*1000))
```

```python
from pyecharts import WordCloud

wordcloud = WordCloud(width=800, height=620)
wordcloud.add("", label[4:], attr[4:], word_size_range=[20, 100])
wordcloud
```

```python
province = data.groupby('Province')['Province'].count()
province
```

```python
from pyecharts import Map

value = list(province)
attr = list(province.index)
map = Map("好友地区分布", width=1200, height=600)
map.add(
    "",
    attr,
    value,
    maptype="china",
    is_visualmap=True,
    visual_text_color="#000",
)
map
```
